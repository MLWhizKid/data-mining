# Data Preprocessing


```python
# 检查缺失值情况，并计算每列的缺失值数量
titanic.isnull().sum()
```

```python
titanic.describe()#生成数据框的统计描述信息。
```

```python
#显示数据框的列名、每列的非空值数量、每列的数据类型以及占用的内存等信息。
titanic.info()
```

```python
# 创建一个热力图，可视化 titanic 数据框中的缺失值情况
sns.heatmap(titanic.isnull(), yticklabels=False, cmap='viridis')
# x y 同时
#sns.heatmap(titanic.isnull(), xticklabels=True, yticklabels=True, cmap='viridis')
```

```python
# 使用中位数填充 'Age' 列中的缺失值。
titanic["Age"].fillna(titanic['Age'].median(), inplace = True)
# 删除 'Cabin' 列。
titanic.drop("Cabin", axis = 1,inplace = True)
# 使用众数填充 'Embarked' 列中的缺失值。
titanic["Embarked"].fillna(titanic)['Embarked'].mode()[0], inplace = True
```
