```python
import pandas as pd #导入pandas库
# 读取名为 "titanic.csv" 的 CSV 文件，并将数据存储为名为 df 的数据框
df = pd.read_csv("titanic.csv")
# 输出 df 数据框的前几行
df.head()
```

## Basic information of the Data
**This provides basic information about the data

```python
df.info#显示数据基本信息
```

```python
df.describe()#生成数据框的统计描述信息。
```

```python

# 计算各列的均值
mean = df.mean()

# 计算各列的中位数
median = df.median()

# 计算各列的众数
mode = df.mode().iloc[0]

# 打印均值
print(mean)

# 打印中位数
print(median)

# 打印众数
print(mode)

```

```python
# 计算 "Age" 列的均值
age_mean = df["Age"].mean()

# 打印均值
print(age_mean)
```

# Data Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 使用 Seaborn 绘制 "Age" 列的直方图，并显示核密度估计（KDE）
sns.histplot(df['Age'], kde=True).set(title='Age')

# 显示图形
plt.show()
```

```python
sns.histplot(df['Fare'],kde = True).set(title = 'Fare')
plt.show()
```

```python
# 使用 Seaborn 绘制箱线图，比较 "Pclass" 列和 "Age" 列之间的关系
sns.boxplot(x='Pclass', y='Age', data=df)

# 显示图形
plt.show()
```

```python
# 计算 "Age" 列和 "Pclass" 列之间的相关系数
corelation_age_fare = df['Age'].corr(df['Pclass'])
##相关系数的取值范围为 -1 到 1，接近 -1 表示负相关，接近 1 表示正相关，接近 0 表示无线性相关。
# 打印相关系数
print(corelation_age_fare)
```

```python
# 计算数据框 df 中各列之间的相关系数
corr = df.corr()

# 打印相关系数矩阵
print(corr)

# 使用 Seaborn 绘制相关系数矩阵的热力图
sns.heatmap(corr, annot=True, cmap='Greens')
#颜色越深表示相关性越强，绿色表示正相关，红色表示负相关，白色表示无相关性。
```
