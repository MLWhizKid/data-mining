```python
housing_df = housing_df.rename(columns={'TOTAL VALUE': 'TOTAL_VALUE'})#把数据列中名称更改
housing_df.columns = [s.strip().replace(' ', '_') for s in housing_df.columns]
#strip() 方法用于去除列名两端的空白字符，replace(' ', '_') 用下划线替换列名中的空格。最终，所有列名都被修改为去除空白字符并用下划线代替空格的形式。
housing_df.head()#返回前五行数据
```

```python
print('Mean of TOTAL_VALUE = ',housing_df['TOTAL_VALUE'].mean())#计算平均值并打印只需更改函数名称
print('Min of TOTAL_VALUE = ',housing_df['TOTAL_VALUE'].min())
```

```python
housing_df.sample(100)#随机抽选100行数据
```

```python
weights = [0.9 if rooms > 10 else 0.1 for rooms in housing_df.ROOMS]
#遍历了数据框 housing_df 中的 'ROOMS' 列的每个元素，将每个元素赋值给变量 rooms。
#根据条件 rooms > 10，如果房间数大于 10，则相应的权重为 0.9；否则，权重为 0.1。
housing_df.sample(100, weights=weights)#（数量，指定了用于抽样的权重列表。）
```

```python
weights = [0.9 if rooms > 10 else 0.1 for rooms in housing_df.ROOMS]
#要查看权重为 0.9 和 0.1 的数量，可以使用 value_counts() 方法来统计列表中各个元素的频数
weights_counts = pd.Series(weights).value_counts()
print(weights_counts)
#pd.Series(weights) 将权重列表转换为 Pandas Series 对象，以便可以使用 value_counts() 方法进行统计。
#value_counts() 方法用于计算 Series 中每个唯一元素的频数。
#weights_counts 是包含权重频数统计结果的 Series 对象。
#print(weights_counts) 将权重的频数打印到控制台。
```
